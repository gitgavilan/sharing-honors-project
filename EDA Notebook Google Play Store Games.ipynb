# How Does Rating and Rating Count Influences the number of installs of a Mobile Game On Google Play Store

The following EDA analyzes a modified data set that contains information related to the rating, rating count, and maximum installs of about 37,000 mobile games that were published on the Google Play Store. The original data set was modified, so it could be more manageable. The original data set contained more than 600,000 applications that were reduced to 37,000 after eliminating applications that were not mobile games, applications that had more than one million downloads, and applications that did not have any type of rating. If you would like to access the original data set, the following link will redirect you to it. [Click Here to see the Original Data Set](https://www.kaggle.com/gauthamp10/google-playstore-apps)

## Information about the Data Set

This project only analyses the effect that Rating and Rating Count has on the number of installs that a mobile game published on the Google Play Store has. In addition, I will also be taking into account the categories of games, so I can see if there is a different relation depending on the category of the mobile game.

- **Rating:** It is an average of all the ratings that an app has. The lowest is 1.0 and the highest is 5.0.
- **Rating Count:** It is the number of ratings that an app has.
- **Maximum Installs:** It is the number of times that an app was download and installed.
- **Category:** It is the type of game. For example, there are games of action, adventure, or puzzles.

### Importing the Necessary Libraries and the Modified Data Set 

```python
import numpy as np
import pandas as pd
import seaborn as sns

apps_data = pd.read_csv("Mobile Game Data Modified.csv")
sns.set(rc = {'figure.figsize':(12,10)})
```

## Data Analysis Using Descriptive Statistics and Informative Graphics

### Rating Count vs Maximum Installs

```python
vis1 = sns.scatterplot(x = 'Rating Count', y = 'Maximum Installs', data = apps_data, hue = 'Category')
```

This is a scatter plot that shows the Rating Cout as the independent variable, and the Maximum Installs as the dependent variable. As can be seen, it does seems to be a strong relationship between these two. As the Rating Count increase, the Maximum Installs increase rapidly. Indeed, we can see that there are more games with a lower Rating Count and higher Maximum Installs than there are games with higher Rating Count and Higher Maximum Installs. However, it seems that there is a stronger relationship between these two variables when we look for the different categories. It can be appreciated that Puzzle Games (blue points) have more installs when they have a bigger Rating Count. On the other hand, Strategy Games (green points) have the same relation but it seems that their installs grow faster when the Rating Count increase. On the whole, a small increase in the Rating Count causes a big increase in Maximum Installs. Nevertheless, there are also a lot of mobile games that have a small Rating Count and lower installs which might indicate that each game category has a different relation between Rating Count and Maximum Installs.

```python
model = sns.lmplot(x = 'Rating Count', y = 'Maximum Installs', data = apps_data, height = 10)
```

By using the lmplot function of the seaborn library, I was able to create the best fit line of the Rating Count vs Maximum Installs graph. This might not be the best data to use this type of method. However, with this, I was able to see that there is an increasing relation between these two variables. In addition, as previously stated this relation might be stronger depending on the category of the mobile game. For this reason, I plotted the best fit line of Rating Counts vs Maximum installs for each game category in the following graph.

```python
model = sns.lmplot(x = 'Rating Count', y = 'Maximum Installs', data = apps_data, height = 10, hue = 'Category')
```

As concluded previously it does seem that each game category has a different relation between the Rating Count and the Maximum installs. Indeed, as stated previously Puzzle games (blue line) increase their Maximum Installs as Rating Count increase faster than Strategy Games (green line). According to this graph, Racing games are the ones that most benefit from the increase of Rating Count. In conclusion, there is an important relation between the Rating Count and the Maximum installs that a mobile game has, and this relation change between game categories; nevertheless, there seem to be other factors affecting the Maximum Installs since there are many games with a big Rating Count, but their installs do not increase that much.

### Rating vs Maximum Installs

```python
vis5 = sns.scatterplot('Rating', 'Maximum Installs', data = apps_data, hue = 'Category')
```

This is a scatter plot where the Rating is the independent variable, and the Maximum Installs is the dependent variable. It can be seen that this plot shows a stronger relationship between the Rating and Maximum Installs than the relationship that exists between the Rating Count and the Maximum Installs. This can be concluded because there are very few mobile games with a lot of installs and a low rating. Indeed, at least more than 60% of mobile games that have a lot of installs have a rating higher than 3.0. In addition, I tried to see if there is a different relationship between each mobile game category; however, I did not find any pattern or relation between the different categories.

```python
model = sns.lmplot(x = 'Rating', y = 'Maximum Installs', data = apps_data, height = 10)
```

This time the best fit line does not shows the real relationship between these two variables as it did in the previous graphs that were analyzed. This is mainly because of the meaning of the line and the meaning of the points. While the line tells us that the rate of change of the Maximum Installs with respect to the Rating is very slow, the points tell us that only the mobile games with a higher rating can achieve a lot of installs. Therefore, the dest fit line does not make any sense in this graph. In conclusion, those mobile games that have a rating bigger than 3.0 are mobile games that are likely to have success. In addition, the relation between Rating and Maximum Installs is not affected by the different categories of games.
